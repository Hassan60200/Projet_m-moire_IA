```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# Charger les données du fichier Excel dans un dataframe
df = pd.read_excel('./data/logement_sociaux_2022.xlsx', sheet_name='DEP', engine='openpyxl')

france = gpd.read_file('./data/departement.geojson')
departements = france[['code', 'nom', 'geometry']]


# Supprimer les deux dernières lignes
df = df.iloc[:-2]

# Extraire les colonnes pertinentes
num_departement = df.loc[4:, 'Logements sociaux au 1er janvier 2022 : comparaisons départementales'].dropna().values
nom_departement = df.loc[4:, 'Unnamed: 1'].dropna().values
logements_sociaux_par_10000_habitants = df.loc[4:, 'Unnamed: 2'].dropna().values
loyer_moyen_par_m2 = df.loc[4:, 'Unnamed: 5'].dropna().values

# Trier les départements par nombre de logements sociaux décroissant
indices_tries = np.argsort(-logements_sociaux_par_10000_habitants)
num_departement_tries = num_departement[indices_tries]
nom_departement_tries = nom_departement[indices_tries]
logements_sociaux_par_10000_habitants_tries = logements_sociaux_par_10000_habitants[indices_tries]
loyer_moyen_par_m2_tries = loyer_moyen_par_m2[indices_tries]

# Afficher le graphique des départements avec le plus grand nombre de logements sociaux
plt.figure(figsize=(10,6))
plt.bar(num_departement_tries[:10], logements_sociaux_par_10000_habitants_tries[:10])
plt.title('Les 10 départements avec le plus de logements sociaux')
plt.xlabel('Départements')
plt.ylabel('Logements sociaux pour 10 000 habitants')
plt.show()

# Trier les départements par loyer moyen par mètre carré décroissant
indices_tries = np.argsort(-loyer_moyen_par_m2)
num_departement_tries = num_departement[indices_tries]
nom_departement_tries = nom_departement[indices_tries]
logements_sociaux_par_10000_habitants_tries = logements_sociaux_par_10000_habitants[indices_tries]
loyer_moyen_par_m2_tries = loyer_moyen_par_m2[indices_tries]
      
# Afficher le graphique des départements avec les loyers moyens les plus élevés
plt.figure(figsize=(10,6))
plt.bar(num_departement_tries[:10], loyer_moyen_par_m2_tries[:10])
plt.title('Les 10 départements avec le loyer moyen le plus élevé')
plt.xlabel('Départements')
plt.ylabel('Loyer moyen par mètre carré')
plt.show()

# Charger les données des contours des départements français
france = gpd.read_file('./data/departement.geojson')
departements = france[['code', 'nom', 'geometry']]

# Ajouter une colonne avec le nombre de logements sociaux par habitant
departements['logements_sociaux_par_habitant'] = 0.0
for i in range(len(num_departement_tries)):
    code_dep = num_departement_tries[i]
    logements_sociaux = logements_sociaux_par_10000_habitants_tries[i] / 10000 * departements.loc[departements['code']==code_dep].shape[0]
    departements.loc[departements['code']==code_dep, 'logements_sociaux_par_habitant'] = logements_sociaux

# Afficher la carte des départements colorés en fonction du nombre de logements sociaux par habitant
fig, ax = plt.subplots(figsize=(10, 6))
departements.plot(column='logements_sociaux_par_habitant', cmap='YlOrRd', linewidth=0.1, ax=ax, edgecolor='black')
plt.title('Nombre de logements sociaux par habitant dans chaque département')
plt.axis('off')

# Ajouter une légende
sm = plt.cm.ScalarMappable(cmap='YlOrRd')
sm.set_array(departements['logements_sociaux_par_habitant'])
cbar = plt.colorbar(sm)

cbar.set_label('Nombre de logements sociaux par habitant')

plt.show()
```
